```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
path='/scrap_price.csv'
data=pd.read_csv(path)
X = data.loc[:, ['carlength',]].values
y = data.loc[:, "price"].values
n_samples,n_featers=X.shape
weights=np.zeros(n_featers)
bias=0
lrnrate=.00001
iterations=36000000

def linerregression(weights,bias,X,y,lrnrate,iterations):
    n_samples,n_featers=X.shape
    for nth_iter in range(iterations):
        y_predict=np.dot(X,weights)+bias
        error=y_predict-y
        dweights=2*np.dot(X.T,error)/n_samples
        dbias=2*np.sum(error)/n_samples
        weights=weights-(lrnrate*dweights)
        bias=bias-(lrnrate*dbias)
        #if nth_iter%5000:
          #print("mse",mse(X,y,weights,bias),nth_iter)
    return weights,bias


def mse(X,y,weights,bias):
    error=np.sum(((np.dot(X,weights)+bias)-y)**2)
    return error


weights,bias=linerregression(weights,bias,X,y,lrnrate,iterations)
print("mse",mse(X,y,weights,bias))
print('weights',weights,'bias',bias)
def predict(X,y,weights,bias):
    y_pred=(np.dot(X,weights)+bias)
    return y_pred
y_pred=predict(X,y,weights,bias)



if X.shape[1] == 1:
    plt.scatter(X, y, color='blue', label='Actual')
    plt.plot(X, y_pred, color='red', label='Prediction')
    plt.xlabel("Feature")
    plt.ylabel("Target")
    plt.legend()
    plt.show()
```
